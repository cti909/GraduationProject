```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor
from sklearn.multioutput import MultiOutputRegressor
from sklearn.metrics import mean_squared_error
```

```python
df=pd.read_csv("D:\Đồ án tốt nghiệp\AI\KNN recommend\dataset_crop.csv")
df.head(5)
```

```python
data = df[['N', 'P', 'K', 'ph', 'label', 'category1', 'category2', 'category3', 'category4', 'category5']]
data
```

```python
data['label'].value_counts()
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.multioutput import MultiOutputRegressor
from sklearn.metrics import mean_squared_error
```

```python
X = data[['N', 'P', 'K', 'ph']]
y = data[['category1', 'category2', 'category3', 'category4', 'category5']]
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
X_scaled
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
model = RandomForestRegressor(n_estimators=100, random_state=42)

# Sử dụng MultiOutputRegressor để huấn luyện mô hình hồi quy đa đầu ra
multi_target_model = MultiOutputRegressor(model)
multi_target_model.fit(X_train, y_train)
```

```python
y_pred = multi_target_model.predict(X_test)

# đánh giá
mse = mean_squared_error(y_test, y_pred)
print(f'Mean Squared Error: {mse}')

print("True values:")
print(y_test.head())

print("Predicted values:")
print(pd.DataFrame(y_pred, columns=y.columns).head())
```

```python
new_data = pd.DataFrame({'N': [86], 'P': [36], 'K': [42], 'ph': [6.5]})

new_data_scaled = scaler.transform(new_data)
predicted_categories = multi_target_model.predict(new_data_scaled)

print("Predicted categories:")
print(pd.DataFrame(predicted_categories, columns=y.columns))
```

```python
from joblib import dump

dump(multi_target_model, 'model/recommend_model.joblib')
```

```python
def get_plant_to_class(index):
    index_to_class = {
        0: 'lúa',
        1: 'chuối',
        2: 'xoài',
        3: 'dưa hấu',
        4: 'cam'
    }
    return index_to_class[index]
```

```python
from joblib import load

loaded_model = load('model/recommend_model.joblib')

check_data = pd.DataFrame({'N': [78], 'P': [41], 'K': [52], 'ph': [7.6]})
check_data_scaled = scaler.transform(check_data)

predicted_categories = loaded_model.predict(check_data_scaled)

print("Predicted categories:")
print(pd.DataFrame(predicted_categories, columns=y.columns))
```

```python
predicted_categories
```

```python
top_3_indices = (-predicted_categories).argsort()[0][:3]
top_3_labels = [get_plant_to_class(idx) for idx in top_3_indices]
top_3_probabilities = predicted_categories[0][top_3_indices]
top_3_indices, top_3_labels,top_3_probabilities
```
